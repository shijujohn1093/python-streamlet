# Date Skewness

```python
from statsmodels.stats.stattools import medcouple
# positive
skew_data = np.array([1,2,3,99])
skewness = medcouple(skew_data)

print(skewness)
```

```python
# Negative skewness
skew_data = np.array([1,99,100,111])
skewness = medcouple(skew_data)

print(skewness)
```

```python
# Normal skewness
skew_data = np.array([1,2,3,4,5])
skewness = medcouple(skew_data)

print(skewness)
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Load the Iris dataset
df = px.data.iris()

# Drop the categorical 'species' column
df_numeric = df.drop(columns=['species'])

# Calculate the correlation matrix
corr_df = df_numeric.corr()

# Create a heatmap using the correlation matrix
fig = go.Figure(data=go.Heatmap(
    z=corr_df.values, 
    x=corr_df.columns, 
    y=corr_df.index, 
    colorscale='Blues'
))

# Display the figure
fig.show()
```
